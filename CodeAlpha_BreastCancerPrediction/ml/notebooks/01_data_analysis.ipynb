# Breast Cancer Prediction - Data Analysis and Exploration

This notebook focuses on loading and exploring the Breast Cancer Wisconsin dataset.

The goal is to understand the structure of the medical data, inspect feature distributions, and analyze relationships between variables before applying any machine learning models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
```

We load the Breast Cancer Wisconsin dataset from scikit-learn.

This dataset contains features computed from digitized images of breast mass and is commonly used for medical classification tasks.

```python
data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name="target")

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

We inspect the dataset to understand feature names, data types, and initial statistics.

```python
X.head()
```

```python
X.info()
```

```python
X.describe().T
```

The target variable represents diagnosis:

0 - Malignant  
1 - Benign

We analyze class distribution to understand dataset balance.

```python
y.value_counts()
```

```python
sns.countplot(x=y)
plt.title("Class Distribution (0 = Malignant, 1 = Benign)")
plt.show()
```

We visualize distributions of selected important features to observe separation patterns between malignant and benign tumors.

```python
features = ["mean radius", "mean texture", "mean perimeter", "mean area"]

plt.figure(figsize=(12,8))
for i, col in enumerate(features):
    plt.subplot(2,2,i+1)
    sns.histplot(data=X, x=col, hue=y, bins=30, kde=True, palette="Set1")
    plt.title(col)
plt.tight_layout()
plt.show()
```

Correlation analysis helps identify relationships between features and detect redundancy.

```python
plt.figure(figsize=(12,10))
sns.heatmap(X.corr(), cmap="coolwarm", linewidths=0.2)
plt.title("Feature Correlation Heatmap")
plt.show()
```

We store the dataset locally to maintain reproducibility across all notebooks and scripts.

```python
df = pd.concat([X, y], axis=1)
df.to_csv("../datasets/breast_cancer.csv", index=False)

print("Dataset saved to ml/datasets/")
```

Conclusions from data analysis:

- The dataset contains 30 numerical features describing tumor characteristics.
- The target variable is binary and relatively well balanced.
- Several features show strong separation between malignant and benign classes.
- High correlation exists between certain features, indicating possible redundancy.

These observations motivate the use of feature scaling, feature selection, and multiple classification models.
